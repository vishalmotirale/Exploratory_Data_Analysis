### Do urban students score higher that rural students ?

```python
df.G1.sum()
```

```python
df.G2.sum()
```

```python
df.G3.sum()
```

```python
sns.kdeplot(df.loc[df['address'] == 'U', 'G1'], label = 'Urban', shade = True)
sns.kdeplot(df.loc[df['address'] == 'R', 'G1'], label = 'Rural', shade = True)
plt.title("Performance comparison for G1")
plt.xlabel("Grade 1")
plt.show()
```

#### The resulting plot allows for an interpretation of the distribution of first grade performance ('G1') for 'Urban' and 'Rural' students.For example, if one KDE curve is shifted to the right or left compared to the other, Here the curve for urban students is shifted to the right, it means that urban students, on average, achieve higher grades in their first-grade performance ('G1') compared to rural students.The shading under the curves provides insight into the relative density of grades within each group.Here the KDE curve for urban students is taller (i.e., has a higher density) compared to the curve for rural students at a first grade level, it suggests that there are more urban students who achieved that first grade level compared to rural students. Overall, the plot facilitates a visual comparison of first grade performance between urban and rural students.

### So,according to kdeplot we can easily interpret that the 'Urban' students scores higher than the 'Rural' students.
